```python
from utils.Backtest import MultipleAssetBacktest
import pymongo
import json
import os
```

```python
backtest = MultipleAssetBacktest()
```

```python
backtest.get_data_and_compute_indicators()
backtest.backtest()
backtest.build_book_of_trade()
# Load book of trade and capture the ID
trade_book_id = backtest.load_book_of_trade()
print(f" Trade Book ID saved: {trade_book_id}")

# Save the trade_book_id to a file so Streamlit can read it
config_file = os.path.join('utils', 'Monitoring', 'trade_book_id.json')
os.makedirs(os.path.dirname(config_file), exist_ok=True)
with open(config_file, 'w') as f:
    json.dump({'trade_book_id': trade_book_id}, f)
print(f" Trade Book ID saved to: {config_file}")
```

```python
print(f" Trade Book ID saved to: {config_file}")
```

```python
# Run Streamlit Monitoring Dashboard
# The trade_book_id from the previous cell will be automatically used by Main.py
import subprocess
import os
import sys

# Get the current working directory (should be the backtest directory)
current_dir = os.getcwd()
# Build path to Monitoring directory (utils/Monitoring from backtest directory)
monitoring_path = os.path.join(current_dir, 'utils', 'Monitoring')
main_py_path = os.path.join(monitoring_path, 'Main.py')

# Check if the file exists
if os.path.exists(main_py_path):
    print(f"Starting Streamlit dashboard from: {main_py_path}")
    print("The dashboard will use the Trade Book ID from the previous backtest execution.")
    print("The dashboard will open in your default browser.")
    print("To stop the server, interrupt this cell (Kernel -> Interrupt)")
    
    # Save original directory
    original_dir = os.getcwd()
    
    # Change to the Monitoring directory to run streamlit
    # This ensures relative imports work correctly
    os.chdir(monitoring_path)
    
    # Run streamlit
    try:
        subprocess.run([sys.executable, "-m", "streamlit", "run", "Main.py"], check=True)
    except KeyboardInterrupt:
        print("\nStreamlit server stopped.")
    except subprocess.CalledProcessError as e:
        print(f"Error running Streamlit: {e}")
        print("Make sure streamlit is installed: pip install streamlit")
    finally:
        # Change back to original directory
        os.chdir(original_dir)
else:
    print(f"Error: Main.py not found at {main_py_path}")
    print(f"Current directory: {current_dir}")
    print(f"Looking for Monitoring at: {monitoring_path}")
    print("Please check the path to the Monitoring directory.")
```
